```python
switch_to => DOM 변경해주는 일
switch_to.window => 창 변경 - 창 마다 current_url 바뀜
switch_to.frame => 같은 창 또 다른 DOM 변경, current_url 안바뀜
          -> 어느 DOM을 해석할지 순서만 바꾸는 것

DOM - DOM(iframe, 돔 안의 돔) => 당장에 눈에 보이는 코드를 숨길 수 있음
```

```python
driver.page_source
# 여기 내용이 없는 이유가 iframe 으로 감싸서 눈에 보이지 않음.
# iframe 을 많이 씀
# 돔 안에 돔이 있을 때는 frame 을 교체해가면서 쓰는 것.
```

```python
dom1_body = driver.find_element(By.CSS_SELECTOR, 'body')
```

```python
driver.switch_to.frame(iframe)
# dom2_body 로 스위치
```

```python
dom2_body = driver.find_element(By.CSS_SELECTOR, 'body')
```

```python
dom1_body, dom2_body
```

```python
dom1_body.tag_name
# 지금 iframe 영역으로 스위치 시켰기 때문에 에러가 나옴.
```

```python
driver.switch_to.default_content()
```

```python
dom2_body.tag_name
# 바로 위에서 돔을 또 스위치 해줬기 때문에 찾아가지 못함
```

```python
# 창 전환, 돔 전환을 잘 알아야 함.
# 그리고 explicit, implicit wait 도 알아놓으면 좋음
#  -> 렌더링 되는 시간이 필요하기 때문에 그려지지 않았다는 둥의 에러가 뜸
#  -> 그래서 explicit wait 이 필요함
```
