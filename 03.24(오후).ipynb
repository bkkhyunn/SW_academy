```python
# Tabulation -> groupby(pandas)
#from functools import
from itertools import groupby
#from operator
```

```python
a = [1,1,1,1,2,2,1,2,3,1,3,3,3]
```

```python
t = groupby(a)
# Init signature: groupby(iterable, key=None)
```

```python
next(t) # 1,1,1,1 -> 2,2 -> 1 -> 2 -> 3 -> 1 -> 3,3,3
```

```python
sorted()
# Signature: sorted(iterable, /, *, key=None, reverse=False)
# 여기서 key 를 함수로 했었음
```

```python
from operator import itemgetter
```

```python
t = groupby(a, key=itemgetter)
```

```python
next(t)
```

```python
t = groupby(a)

for i,j in t:
    print(dir(j))
# print 해서 <itertools._grouper at 0x7f87ba5249a0> 것들이 안나옴 => groupby로 next를 다 해서(그래서 합쳐놔야 함.)
```

```python
# groupby 했던 것을 next를 할 수 있게 되어있음. 즉 <itertools._grouper at 0x7f87ba5249a0> 도 iterator, generator 와 같음

t = groupby(a)

for i,j in t:
    print('++++')
    for b in j:
        print('-----')
        print(b)
```
