#### 머신러닝 가정

```python
# 머신러닝은 수많은 가정이 들어감. 내가 가진 데이터로부터 미지의 안보는 세상을 예측하는 것.
# 이런 것은 귀납법.
# 귀납법은 하나하나의 개별적인 사실로부터 일반적인 사실을 도출하는 방법.
# 머신러닝은 전세계 iris 에 관한 모든 데이터를 가질 수 없기 때문에, 그 중에서 일부만 가지고 있음.(데이터); 이게 통계의 기본
# 전체 모집단에서 일부 추출된 걸로, 전체를 예측하는 것. 이게 기본이다!
# 그려려면 여러 조건이 필요함
# 첫째로 내가 가진 데이터가 전체를 잘 반영해야 함. '대표성'; 대표성이 없는 데이터를 편향(bias)되어 있다고 이야기함. 대표한다고 가정하는 것.
# 우리는 데이터 대표성이 있다고 믿는다, 가정한다.
# 데이터가 많으면 많을수록 큰 수의 법칙(수가 자기의 분포에 맞게 됨)을 따름
# 그 데이터의 일부를 모델 만들 때 학습 데이터로 만들고, 일부는 정답 데이터로 모델이 맞는지 안 맞는지 테스트.
# 즉, training, test set로 분리 시킴. 그것을 hold-out 이라고 함.
# hold-out의 의미는 내가 가진 데이터가 대표성을 가지고 있다고 믿는 것.
```

```python
sum(lr.predict(X_test) == y_test), sum(knn.predict(X_test) == y_test)
```

```python
# hold-out 에서 문제점이 있음. 랜덤하게 쪼갬. 즉, 특정한 데이터에 따라서 성능이 달라질 수 있음.
# 쪼개는 것을 다시
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-1], iris.target)
lr.fit(X_train, y_train)
knn.fit(X_train, y_train)
sum(lr.predict(X_test) == y_test), sum(knn.predict(X_test) == y_test)
```

```python
# 최종 서비스할 때는 다시 합쳐서 재학습을 시킴; 데이터가 많으면 많을수록 좋기 때문에.
# 즉 처음에는 이미 가진 데이터로
# 존재하는 general 한 알고리즘을 가지고 내부적인 값을 바꿔서
# 내 데이터 전용으로 맞춰가는 것.
# 그 맞춰가는 과정을 "학습" 즉 "머신러닝"이라고 함
```

```python
# 여기까지 한 것이 뼈대.
# 데이터를 쪼개지 않으면 객관적인 성능 측정을 하지 못함. 그러나 내가 가진 데이터는 대표성을 띈다고 가정하기 때문에,
# 머신러닝에서 가장 어려운 주제 중에 하나는, 머신에 의해서 학습된 모델의 절대적인 성능.
# 절대적인 성능을 파악하기가 불가능함 => 내가 가진 데이터가 대표성을 가지지 않기 때문.
```

```python
# 특정 문제에 대해서 항상 성능 좋은 알고리즘 없음. 그래서 이것저것 다 해보고 가장 성능 좋은 것을 찾아야 함.
# 또는 만들어야 함.
# 절대적 성능 파악할 수 없음. 내가 가진 데이터가 전체가 아니고 일부이기 때문.
```

```python
# hold-out; 랜덤으로 쪼갬
# 이 때 문제가 극단적으로 쪼개질 수도 있음. 따라서 큰 데이터를 사용하면 문제점이 좀 줄어듦
# 그러넫 큰 데이터를 사용하면 시간이 많이 걸림.
# 작은 데이터셋을 쓸 때 쓰는 테크닉이 있음.
```

```python
# 이상치(outlier)를 제거하는 것을 노이즈 제거 라고 한다.
# boxplot 에서 말하는 노이즈가 절대적이지는 않음. 그래서 없이 학습해보고, 있이 학습해보고 해야 함.
# noise 인 애들을 필터링할 줄 알아야함.
```

```python
# map 을 활용해서 outlier detection 을 해보자
def find_outlier(df):
    IQR = df.quantile(0.75) - df.quantile(0.25)
    temp = df[(df.quantile(0.75) + 1.5*IQR < df)|(df.quantile(0.25) - 1.5*IQR > df)]
    return temp.dropna(how='all').index
```

```python
find_outlier(iris)
# 조건에 안맞으면 NaN 나옴
```

```python
find_outlier(iris).dropna(how='all')
# how : {'any', 'all'}, default 'any'
#     Determine if row or column is removed from DataFrame, when we have
#     at least one NA or all NA.

#     * 'any' : If any NA values are present, drop that row or column.
#     * 'all' : If all values are NA, drop that row or column.
```
