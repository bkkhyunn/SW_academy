```python
# tokenizer 봤는데 의미 없는 것도 있었고 중복된 것들도 있었음.
# 어간추출(Stem); 용언 + 활용 => 동사나 형용사 뒤에 붙는 것들. ed, ing, ies,... => 어차피 의미는 다 같음. 그런데 표현된 게 달라서
#              다 개별 단어로 표현되어 있을 것. 그래서 뒤의 것 자르고 어간만 쓰자! 그러면 feature 수 더 줄일 수 있어!
#              우리 말로 하면 하다 + 였고, 였으며, 였었고, ... => 줄이고 어간만 찾자!
# 원형추출(Lemma); 원래의 의미를 갖는 원형. 가령 be 동사 -> 문장, 시제, 주어에 따라 be, is, are, ... 여러가지로 쓰임 그런데 다 be 동사
#              원형을 찾아 다 be로 바꾸고 줄이는 것
#              우리 말은 탈락규칙. 원래의 표기들을 바꾸는 애들

# 우리글은 형태소 분석할 때 저 과정들이 다 들어가 있음.
# ex. 할래? 하다 + 청유형 어미 => 원래 표현하는 방식이랑 많이 달라짐. 이럴 때 원형으로 복원하는 것을 원형추출
```

```python
threshold = .3 # 실험을 통해 밝혀야 하는 값인데 일반적으로 .3
cp = 0.0
mft = list()

for t in tokens:
    if cp > threshold:
        break
    cp += text2.vocab().freq(t[0]) # 비율
    mft.append(t)
    #t[1] / text2.vocab().N() # 전체 비율
```

```python
mft
```

```python
len(mft), sum([t[1] for t in mft]), sum([t[1] for t in mft])/text2.vocab().N()
# 14개의 단어가 30프로를 차지함
# 이들을 제거해버리면 전체 개수가 확 줄어버림.
```

```python
threshold = .3 # 실험을 통해 밝혀야 하는 값인데 일반적으로 .3
cp = 0.0
mrt = list()

for t in tokens[::-1]:
    if cp > threshold:
        break
    cp += text2.vocab().freq(t[0]) # 비율
    mrt.append(t)
    #t[1] / text2.vocab().N() # 전체 비율
```

```python
# Heaps 의 법칙
# 사용자 평 하나하나가 문서. 뉴스 분석에서는 뉴스 단 건 하나하나가 문서가 될 것
# 문서가 늘어나면 늘어날수록 거기서 사용되는 단어들도 늘어남.
# 문서를 수집할 때마다 유니크한 단어가 기하급수적으로 증가할 것.
# 그걸 어떻게 감안할 것인지 내용이 힙스의 법칙
# 문서 안의 유니크한 단어(피쳐가 될 것들)들이 무한정 늘어나지 않는다는 것. 특정한 법칙을 띄고 있음. 로그함수같이
# 아무리 수집하고 분석해도 우리가 처리해야하는 유니크한 단어의 수는 어느정도 고정되어 있고 결정되어 있음을 나타내는 법칙
# 문서를 여러개 열어야 증명할 수 있음
# 이는 KoNLPY 로 한글을 이용해서 할 것.
```
