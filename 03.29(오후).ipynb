#### Encoding

```python
# 두번째로,
# 숫자 아닌 데이터가 없으면 안됨
# Encoding 은 feature-extraction 안의 한가지 방법.
# 다양한 형태의 데이터를 수치화하는 것을 feature-extraction 이라고 함.
# 그 중 하나인 Encoding은 문자를 숫자로 바꾸는 것.
```

```python
# 데이터의 4가지 분류
# 범주형(nominal, ordinal)
# 수치형
# 이산형
# 연속형
# 측정의 정도에 따라 interval, ratio 까지

# 사람이 하는 테크닉은 2가지

# 가능한 데이터가 가진 성질 그대로 변화시키는 것이 좋음.
# 예를 들어 남녀는 순서가 있는 데이터가 아님. binary 라고 해서 0과 1로 => dummy (통계쪽)
# 컴퓨터 쪽에서는 One-Hot encoding 이라고 부름. 이름만 있는 애들은 순서가 없음.
# multi-hot 도 있음. 1이 여러개. hot은 1을 뜻함.
# 이걸 왜 쓸까? ; 크기가 무조건 1임. 각각 다른 숫자를 부여하면, 크기와 순서를 가지게 되버림. 따라서 내가 정한 숫자에 따라 모델이 변형될 수도 있
# 음. 따라서 nominal 에서는 더미변수, 원핫 인코딩을 씀. 

# ordinal 에서는 label encoding 이라고 함.

# 딥러닝에서는 내가 이렇게 하지 않고 학습을 통해서 함.
# 단순 원핫 인코딩은 변수마다 관계성을 잃어버릴 수도 있음. 그래서 word embedding 을 함.
```
