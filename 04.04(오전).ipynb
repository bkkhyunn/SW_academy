```python
# 어떠한 특정 요소의 한계 때문에 전체가 제한받는 상황을 병(bottle)의 목(neck)에 비유해서 병목현상(bottleneck effect)이라고 함.
# 즉 일부분에 의한 전체의 하향 평준화.
```

```python
# 비효율적
from os import listdir

def fileids(path):
    fileList = list()
    
    path = path + ('' if path[-1] =='/' else '/')
    
    for f in listdir(path):
        if f.endswith('.txt'):
            fileList.append(path+f)
            
    return fileList
```

```python
D = list()
for file in fileids('news_naver'):
    with open(file, 'r') as fp:
        D.append(fp.read())
```

```python
# complextiy 보자
# space: len(D) * len(v)   -> 줄여야하는 과제가 있음
# time: |Q| * D(전체 문서집합) * |D|   -> 병목 부분을 줄여야 하는 것도 과제
# 이 과제들을 어떻게 해결?
# linked list / vector space model 들이 각각 해결해줄 것.
```

```python
# Complexity
# Space: len(D) * len(V) - 줄여야함 과제 -> Linked List!
# Time: |Q| * D * |D| - 병목부분 줄여야함 과제 -> Vector Space Model!
```

```python
Q = '등이 부부도 성장하면'.split()  # |Q| = 3
result = list()
for q in Q:
    for d in D:   # Bottle neck
        for t in d:
            if q == t:
                result.append(D.index(d))
                break
                
list(set(result))

# OR 연산으로 만든 Boolean 검색
```

```python
# 만약 AND 연산으로 만들고 싶으면 
Q = '윤석열 대통령 방문'.split()
result = list()
for q in Q:
    result.append(list())
    for d in D:   # Bottle neck
        for t in d:
            if q == t:
                result[-1].append(D.index(d))
                break

list(set(result[0]).intersection(set(result[1])).intersection(set(result[2])))
```
