```python
# 해결책은 매우 간단했음.
# perceptron 을 두 개 붙히고 두 개를 다시 붙혔음.
# stacking.
# layer 개념이 나옴. -> 같은 데이터를 받는 것.
# 그래서 이 때부터 input layer, hidden layer, output layer
# hidden layer 가 핵심이 됨. -> 히든 레이어를 어떻게 구성하느냐에 따라서 성능이 완전히 달라짐
```

```python
# 그러나 학습시키는 방법을 찾지 못했음.
# hebbian learning -> perceptron learning -> delta rule -> gradient descent
# 이 해결책의 이름이 바로 back propagation 역전파.
# 미분 중에서도 편미분이 사용됨
```

```python
# Multi Layer Perceptron 이 바로 위 모델. = ANN (Artificial Nenural Network)
# 그 중에서도 hidden layer 가 많은 모델을 딥러닝이라고 함.
```

```python
# input layer 는 신경 안쓰기 때문에 무조건 input 의 개수를 뺌
# 또한 output 도 항상 있는 거라서 1-layer 라고 부르기도 함.
```

```python
# layer 를 차원의 개념으로 보면 됨. layer 가 많으면 많을 수록 복잡한 차원이라서 뉘앙스가 풍부함.
# layer 가 많다는 것을 deep 이라고 하고, layer 가 작은 것을 shallow 라고 함.
# 3개 이상의 layer 를 가진 모델을 딥러닝모델.
# 얘네는
# 인풋값이 들어와서 웨이트와 곱해져 결과값이 나옴. 이 때 학습을 통해서 w 와 b 찾음.
# w와 b 찾고 새로운 결과값을 냄.
# 이 결과값을 내는 과정을 앞쪽으로 값을 주면서 나아간다고 해서 feed-forward 라고 함.
# feed-forward = prediction(예측)이 됨. 
```

```python
# 이게 왜 성능이 좋을까? 라는 의문이 생김.
# linear 모델을 stacking 했는데 왜 성능이 좋을까?
# f(x) = ax + b
# f(f(x)) = a(ax+b)+b
#         = a^2x + ab + b
# 이게 stacking 인데 왜 성능이 좋아지나?
# threshold func => activation func
# 합성함수 중간에 activation func 이 있는데 이 함수가 Non-linear
# linear -> non-linear -> linear
# 즉 중간중간에 activation func 즉 non-linear 함수가 들어가서 성능이 좋아지는 것. 즉 아래 그림에서 perceptron 에서 out 부분
# 모델의 non-linearity
```

```python
# nonlinearity 가 섞여 있어서 성능이 좋아지는 것이고 그게 핵심.
# linear combination 했는데 성능이 좋은 이유임.
# 이 activation func 이 non-linear func. -> 미분 불가 함수. -> 미분을 통해서 최소값 못 찾음.
# 그래서 이게 sigmoid(logistic regression) 로 바뀜. sigmoid 함수는 non-linear func임.
```

```python
# non-linear 함수가 의미하는 바를 보자.
# colah 홈페이지
# non-linear 함수가 해주는 역할을 activation func 에서 해줌.
# 중요한 구성요소
# perceptron 에서 같은 걸 받으면 같은 layer
# perceptron 별로 activation func 이 있다.
# stacking 된 것이 network
```

```python
# lazy 의 반대말은 eager
# 하나의 데이터에 몇 개 들어오느냐가 중요
# lazy는 이상한 값이 들어오는 걸막을 순 없는데
# eager는 막을 수 있다ㅏ
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(6, input_shape=(6,)),
    tf.keras.layers.Dense(5)
])
```

```python
model.summary()
# None 은 어느 수나 상관없다는 것.
# 이것은 결과의 모양
# weight 는 함수
# input을 weight 곱해서 결과 나오게 하는 것이 퍼셉트론
```

```python
model.weights
# weights 는 variable 로 구성되어 있음. 자기 자신이 나중에 바뀔 것이기 때문에
```

```python
# 구해야 할 가중치가 많다는 것은 구해야 할 차원이 많음.
# 내가 가진 데이터로 가능하게 복잡하게 만드는 것은 안 좋음. - 오캄의 면도날
# 그 알고리즘은 결국 우리가 짜는 것.
# 퍼셉트론 많으면 많을수록 성능이 좋음. 많으면 오버피팅 할 수도
model = tf.keras.models.Sequential(
    tf.keras.layers.Dense(6, input_shape=(4,)),
    tf.keras.layers.Dense(3)  # iris data 결과가 3가지니까 3가지로 나와야 분류 -> 분류에서는 가능한 맞춰준다!

)
# NN 에서는 y 값을 무조건 one-hot encoding. 결과를 컬럼 개수와 똑같이 맞추기 위해서.
```
