```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
# Lasso 는 linear regression 에 regularizer 추가. 절대값
#  (1 / (2 * n_samples)) * ||y - Xw||^2_2 + alpha * ||w||_1
# 최솟값을 찾기 위해서 누군가가 하나 더 딸려 있어서 나의 행동반경이 제약을 받음. -> 오버피팅 할 때 그 데이터에 딱 맞추지 않고 여유를 두는
# 오버피팅 막는 테크닉. L1 = Lasso
Lasso
# Init signature:
# Lasso(
#     alpha=1.0,
#     *,
#     fit_intercept=True,
#     precompute=False,
#     copy_X=True,
#     max_iter=1000,
#     tol=0.0001,
#     warm_start=False,
#     positive=False,
#     random_state=None,
#     selection='cyclic',
# )
```

```python
# L2 = Ridge. 제곱
# ||y - Xw||^2_2 + alpha * ||w||^2_2
# 딥러닝에서는 L2 를 Weight Decay. 정확하게 따라가지 않고 느슨하게 따라가도록 하기 때문
# cs 에서는 L2 가 더 많이 쓰임.
Ridge
# Init signature:
# Ridge(
#     alpha=1.0,
#     *,
#     fit_intercept=True,
#     copy_X=True,
#     max_iter=None,
#     tol=0.0001,
#     solver='auto',
#     positive=False,
#     random_state=None,
# )
```

```python
# 1 / (2 * n_samples) * ||y - Xw||^2_2
# + alpha * l1_ratio * ||w||_1
# + 0.5 * alpha * (1 - l1_ratio) * ||w||^2_2
# L1, L2 같이
ElasticNet
```

```python
# 우리의 첫번째 오버라이딩 MyC
# Tensorflow의 고급기술은 layer 도 오버라이딩 해서 만들 수 있음.
class MyDenseLayer(tf.keras.layers.Layer):
  def __init__(self, num_outputs):
    super(MyDenseLayer, self).__init__()
    self.num_outputs = num_outputs

  def build(self, input_shape):
    self.kernel = self.add_weight("kernel",
                                  shape=[int(input_shape[-1]),
                                         self.num_outputs])

  def call(self, inputs):
    return tf.matmul(inputs, self.kernel)

layer = MyDenseLayer(10)
```
