```python
# Seq2Seq
# 입력도 시퀀스, 출력도 시퀀스
# encoder 는 들어온 문장의 단어를 embedding vector 로 만들고
# 그 단어를 잘 표현하는 context vector를 decoder 에 전달(초기파라미터로)
# 학습이 인코더와 디코더 연결되어 일어남.
# 그 인코더와 디코더에 LSTM 을 씀.
# 디코더에서는 들어온 context vector 에 따라 하나씩 생성함.
# 디코더 학습은 teacher forcing. 강제로 넣어준다는 것.
# 예측할 때는 auto regressive 방식으로 처리함.
```

```python
# Dense layer 가 어휘사전에 있는 단어들의 모든 확률 제시 가능. 다 더하면 1 가능
# softmax 를 쓰면 스코어 값으로 만들어낼 수 있음.
# 학습시킬 때 <sos> 다음에 je 가 가장 높게 나오도록 학습시킨다는 것
# (teacher forcing)
```
