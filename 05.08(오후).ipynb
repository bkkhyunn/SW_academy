```python
위치에 따라 equivalance
다른 형태로 표현했는데, 위치만 변경되는 것
위치를 바꾸면 내가 나타내는 결과, 함수로 생각하면
어떤 특정 함수의 input 의 위치를 바꿨을 떄 다른 결과는 다 똑같고 위치만 달라지면
위치에 따른 equivalance 가 있다고 함.

invariance 는,
어떤 함수에 뭐를 집어넣었을 때 2,
위치를 바꿔도 2.
위치를 바꿔도 항상 결과값이 같으면 위치에 대해서 invariance 하다고 표현함.

즉 전자는 위치바꾸면 위치말고 다른 건 다 똑같은 것(equivalance)
후자는 위치바꾸면, 즉 위치를 바꿔도 결과값이 똑같은 것(invariance)

이 개념을 왜 이야기 했을까?
```

```python
두번쨰 해결책은 window 가 움직이는데,
위치가 바뀌어도 결과값 똑같이 나오는 모델을 만들면 좋음. 이런 이동을 translation.
이동에 따른 invariance 한 모델을 만드는 것이 성능이 아주 좋을 것.

translation 에 대한 Invariance 한 모델을 만드는 것이 목적
위치에 상관없이 항상 똑같은 결과를 나오게 하는 것.
=> 이게 얀 누쿤이 생각한 것. 위치에 상관없는 모델을 만들자!!!

translational invariance 한 모델을 만들자! => 난제가 있음.
```

```python
결과적으로 convolution 할 건데,
이 convolution 이 바로 translational invariance. 위치가 바뀌어도 결과값이 바
뀌지 않는 것.
```

```python
# 신호는 1차
# convolution 이 신호처리(signal)에서 왔음을 알 수 있음.
from scipy.signal import correlate, convolve, convolve2d
import numpy as np
```

```python
a = np.array([1,2,3,4,5,])
```

```python
b = np.array([1,0,1])
```

```python
correlate(a, b)
```

```python
correlate(a, b, mode='same')
```

```python
correlate(a, b, mode='valid')
```

```python
correlate(a, b, mode='full')
```

```python
# 어떻게 계산해서 위처럼 나왔을까?

# mode : str {'full', 'valid', 'same'}, optional
#     A string indicating the size of the output:

#     ``full``
#        The output is the full discrete linear cross-correlation
#        of the inputs. (Default)
#     ``valid``
#        The output consists only of those elements that do not
#        rely on the zero-padding. In 'valid' mode, either `in1` or `in2`
#        must be at least as large as the other in every dimension.
#     ``same``
#        The output is the same size as `in1`, centered
#        with respect to the 'full' output.
```
