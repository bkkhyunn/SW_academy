```python
세타에 대해서 미분. 근데 세타가 x세타. 벡터벡터임. 그래서 세타를 편미분을 해줌.
로그를 미분하면 분수가 됨.
세번째 식에서 괄호 안에가 로지스틱 func 모양임. x가 given 일 때 y의 확률. likelihood 임.
맨 마지막을 0으로 두면 최적화된 세타가 됨.

맨 마지막에서 다시 P 값을 만들어줘야 하는데 세타가 given으로 들어감. 그래서 미분식 만으로는 아무리 하려고 해도 만들어질 수 없음.
그래서 가장 적합한 argmax 한 세타를 단순히 세타를 미분해서는 못찾음. 그래서 그것을 estimation 하는 형태로 가야함.
그게 gradient descent 가 됨.

로지스틱 regression 에서는 그런 형태가 안 만들어져서 estimation 하는 형태로 가야하고 그것을 gradient descent 로 이어진 것.
활성화 함수를 거쳐 다음 레이어로 보내게 됨.
미분을 취해서 최적화된 세타, NN 에서는 w 를 찾아야 하는데 이게 단번에 Linear regression 처럼 찾아지지 않으니 estimation 해서 기울기가
작아지는 쪽으로 추정하자!

NN에서 기본은 에러를 가지고 최소화되는 지점을 찾는게 기본이기 때문에 gradient descent 를 사용하는 것.
```

```python
그래서 오늘 기억하는 것은
확률을 가지고 주어진 데이터셋에서 관측값이 최대화되는 지점을 likelihood 를 추정함.
이떄 단순히 데이터셋만이 아니라 세타라는 파라미터를 선형결합의 형태로 섞음. 단지 그러면 선형 regression 으로 끝남.
여기서는 확률로 표현해서 p 라고 했고, 그 p를 로지스틱으로 구성.
그래서 로지스틱 리그레션에서 사용하는 로지스틱 함수를 이용한 로지스틱 분류가 나타나게 된 것.
내일은 gradient descent 를 활용해볼 것.
```
