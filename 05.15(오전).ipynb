```python
X = np.random.rand(100)*np.pi*2
Y = np.sin(X)

nX = np.c_[np.ones(len(X)), X]
W = np.linalg.inv(nX.T.dot(nX)).dot(nX.T).dot(Y)

plt.scatter(X, Y)
plt.scatter(X, nX.dot(W))

# 차수 증가 -> 더 많은 파라미터 증가, 즉 모델의 complexity 가 높아짐. 이렇게 되면 데이터에 더 fitting 을 잘 시킴. 오버피팅 가능성 높아짐.
#            앞으로 올 데이터에 대해서 잘 못맞춰서 bias 값이 높아질 것.
# 차수 감소 -> complexity 낮아짐. 앞으로 올 데이터에 대해서 성능을 어느정도 보장해줌. 그러나 under-fitting 의 문제가 있음. variance가
#            높아짐.
# 그래서 여기서 얘기하는 것이 occam's razor. 유사한 성능을 낸다면, 모델의 복잡도가 낮은 게 더 좋고 그것을 채택하는 것이 좋다!
# ax + b, ax^2+bx+c
```

```python
X = np.random.rand(100)*np.pi*2
Y = np.sin(X)

nX = np.c_[np.ones(len(X)), X, np.power(X,2), np.power(X,3)]
W = np.linalg.inv(nX.T.dot(nX)).dot(nX.T).dot(Y)

plt.scatter(X, Y)
plt.scatter(X, nX.dot(W))

# complexity 가 높아짐. 데이터에 엄청 잘 fitting 됨. 
# 앞으로 올 데이터에 대해서도 어느정도 유사한 값을 내게끔 우리가 정해줘야 함.
```

```python
from math import exp, factorial

# f 를 n 번 미분. f = exp(몇번 미분하든 자기 자신)
def taylor(x, a=0, N=5):
    # 누적하면 estimation 한 결과
    rst = 0.0
    for n in range(N):
        rst += (exp(a)/factorial(n))*(x-a)**n
    return rst
```

```python
X = np.arange(1,5, .1)
Y = np.exp(X)
tY = list(map(taylor, X))
plt.plot(X, Y)
plt.plot(X, tY) # N 을 늘릴수록 exp 함수에 가까워질 것.

# 그러면, 테일러 급수처럼 특정 점에서 무한대로 미분 가능하고 무한합으로 표현됨을 이해한다면,
# 우리에게 필요한 것은 미분값이고, 미분한 만큼 움직인다면
# 마치 N을 여러번 돌린 것처럼 exp 함수랑 유사해지겠구나! 우리가 찾아야하는 점에 가까워지겠구나를 이해.
# => 그래서 gradient 방법을 적용할 수 있음! iteration 을 여러번 하는 것은 근사치를 잘 만들어내기 위해서 필요한 것임을 잘 이해.
```

```python
def rosenbrock(x1, x2):
    return (1-x1)**2 + 100 * (x2-x1**2)**2

def x1_derivative(x1, x2):
    return -2*(1-x1)-400*(x2-x1**2)*x1

def x2_derivative(x1, x2):
    return 200*(x2-x1**2)
```

```python
from math import sqrt
X = [-1.3, 0.9]
rosenbrock(*X)
u = [x1_derivative(*X), x2_derivative(*X)]
u = [u[0]/sqrt(u[0]**2+u[1]**2), u[1]/sqrt(u[0]**2+u[1]**2)]
```
