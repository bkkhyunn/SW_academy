```python
범용적으로 만들어진 것을 미세하게 튜닝하는 것이 fine tuning.
왜 fine 이라는 말이 붙었을까?
여기서 알아야 하는 것이 catastropy forgetting 현상; NN은 예전에 배운 거 새로 학습시키면 weight 가 달라져서 다 까먹음.
weight 가 크게 바뀌면 예전의 능력까지 다 잃어버리기 때문에(catastropy forgetting) 미세하게 튜닝.
```

```python
1. backbone 을 아예 고정시키고, 백본에서 나온 별 중요하지 않은 애들만 웨이트를 변경하는 경우. 백본으로 나온 결과를 조금 바꾸는 것.
2. 전체를 재학습시키는데, weight 가 바뀌어버리면 옛날 거를 다 까먹어 버리니까 learning rate 를 굉장히 줄여서 아주 미세하게만 바꿈.

상황에 따라 이 두가지 테크닉을 씀.
기존에 남이 있는 것을 가져다 쓰면 우리 데이터로 절대 잘 안됨. 그리고 모든 애들이 내가 가진 데이터가 전체를 대표할 수 없음. 그래서 머신러닝은 항상
문제가 있음. 우리 Task 에 맞추는 것이 아주 중요함.
```
